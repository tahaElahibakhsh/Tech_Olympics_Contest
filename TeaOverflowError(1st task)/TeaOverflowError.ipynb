```python
# TODO:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_tr = pd.read_csv('Data/train.csv').drop(columns='id')
df_ts = pd.read_csv('Data/test.csv').drop(columns='id')
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
        در این سوال شما می‌توانید از هر تکنیک پیش‌پردازش/مهندسی ویژگی دلخواهتان، استفاده کنید.
    <br>
    تکنیک‌هایی که استفاده می‌کنید به شکل مستقیم مورد ارزیابی توسط سامانه داوری قرار <b>نمی‌گیرند.</b> بلکه همه آن‌ها در دقت مدل شما تاثیر خواهند گذاشت؛ بنابراین هر چه پیش‌پردازش/مهندسی ویژگی بهتری انجام دهید تا دقت مدل بهبود پیدا کند، امتیاز بیشتری از این سوال کسب خواهید کرد.
    در این قسمت شما می‌توانید بخشی از داده‌ی موجود را برای اعتبارسنجی در نظر بگیرید.
</font>
</p>

```python
df_tr['store_and_fwd_flag'] = df_tr['store_and_fwd_flag'].map({'N':0, 'Y':1})
df_ts['store_and_fwd_flag'] = df_ts['store_and_fwd_flag'].map({'N':0, 'Y':1})
```

```python
df_tr.drop(df_tr[(df_tr['passenger_count'] == 0) | (df_tr['passenger_count'] > 6)].index, inplace=True)

df_tr.drop(df_tr[df_tr['trip_duration'] > 8000].index, inplace=True)

lon_min, lon_max = -74.05, -73.75
lat_min, lat_max = 40.63, 40.85

df_tr = df_tr[
    (df_tr['pickup_longitude'].between(lon_min, lon_max)) &
    (df_tr['pickup_latitude'].between(lat_min, lat_max)) &
    (df_tr['dropoff_longitude'].between(lon_min, lon_max)) &
    (df_tr['dropoff_latitude'].between(lat_min, lat_max))
].copy()

```

```python
df_ts
```

```python
for df in [df_tr, df_ts]:
    df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
    df['dropoff_datetime'] = pd.to_datetime(df['dropoff_datetime'])

    # ---- pickup features ----
    df['pickup_month'] = df['pickup_datetime'].dt.month
    df['pickup_weekday'] = df['pickup_datetime'].dt.weekday
    pickup_hour_float = df['pickup_datetime'].dt.hour + df['pickup_datetime'].dt.minute / 60
    df['pickup_time_sin'] = np.sin(2 * np.pi * pickup_hour_float / 24)
    df['pickup_time_cos'] = np.cos(2 * np.pi * pickup_hour_float / 24)
    df['pickup_hour'] = df['pickup_datetime'].dt.hour + df['pickup_datetime'].dt.minute / 60


    # ---- dropoff features ----
    df['dropoff_weekday'] = df['dropoff_datetime'].dt.weekday
    dropoff_hour_float = df['dropoff_datetime'].dt.hour + df['dropoff_datetime'].dt.minute / 60
    df['dropoff_time_sin'] = np.sin(2 * np.pi * dropoff_hour_float / 24)
    df['dropoff_time_cos'] = np.cos(2 * np.pi * dropoff_hour_float / 24)
    df['dropoff_hour'] = df['dropoff_datetime'].dt.hour + df['dropoff_datetime'].dt.minute / 60


    df.drop(columns=['pickup_datetime', 'dropoff_datetime'], inplace=True)
```

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    dphi = np.radians(lat2 - lat1)
    dlambda = np.radians(lon2 - lon1)
    a = np.sin(dphi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(dlambda/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

df_tr['distance_km'] = haversine(df_tr['pickup_latitude'], df_tr['pickup_longitude'],
                                 df_tr['dropoff_latitude'], df_tr['dropoff_longitude'])

df_ts['distance_km'] = haversine(df_ts['pickup_latitude'], df_ts['pickup_longitude'],
                                 df_ts['dropoff_latitude'], df_ts['dropoff_longitude'])

```

```python
df_tr['trip_duration_hr'] = df_tr['trip_duration'] / 3600
df_tr['speed_kmh'] = df_tr['distance_km'] / df_tr['trip_duration_hr']
df_tr.drop(columns=['trip_duration_hr'], inplace=True)

df_ts['trip_duration_hr'] = df_ts['trip_duration'] / 3600
df_ts['speed_kmh'] = df_ts['distance_km'] / df_ts['trip_duration_hr']
df_ts.drop(columns=['trip_duration_hr'], inplace=True)
```

```python
df_tr = df_tr[(df_tr['distance_km'] > 0) & (df_tr['distance_km'] < 100)]
df_tr = df_tr[(df_tr['speed_kmh'] > 0) & (df_tr['speed_kmh'] < 120)]
```
